````python
# Pongo returns con los resultados matematicos por si esto se quiere utilizar para otras cosas (almacenandolo en una variable)

def cuantos_pasteles(receta, ingredientes):
    lista = []
    for i in receta.keys():               
        if i in ingredientes.keys():
            print(f'Hay {i}')
            lista.append(i)
        else: 
            print(f'Te falta {i}')

    if len(lista) == len(list(receta.keys())):

            cont = 0                   
            cantidad = round(ingredientes[lista[cont]] / receta[lista[cont]])
            cantidades = []
            while cont < len(lista):
                
                if cantidad >= 1:
                    print(f'Tienes suficiente {lista[cont]}')
                    cantidades.append(cantidad)
                    cont += 1
                
                else:
                    if ingredientes[lista[cont]] < receta[lista[cont]]:
                        print(f'Necesitas {receta[lista[cont]] - ingredientes[lista[cont]]} más de {lista[cont]}')
                        cont += 1
            
            print(f'Puedes hacer {round(sum(cantidades) / len(cantidades))} pasteles')
            return round(sum(cantidades) / len(cantidades))
    else:
        print('Cuando tengas todos los ingredientes ¡nos ponemos a cocinar!')
        return 0
````

````python
def cuantos_pasteles(receta, ingredientes):
    lista = []
    for i in receta.keys():               
        if i in ingredientes.keys():
            print(f'Hay {i}')
            lista.append(i)
        else: 
            print(f'Te falta {i}')

    if len(lista) == len(list(receta.keys())):

        cont = 0   
        cantidades = []                
        while cont < len(lista):
            cantidad = int(ingredientes[lista[cont]] / receta[lista[cont]])  #tiene que estar dentro del loop para que se calcule la cantidad cada vez que cambia el contador, sin solo se define la variable una vez

            if cantidad >= 1:
                print(f'Tienes suficiente {lista[cont]}')
                cantidades.append(cantidad)
                cont += 1
        
            else:
                print(f'Necesitas {receta[lista[cont]] - ingredientes[lista[cont]]} más de {lista[cont]}')  
                cantidades.append(cantidad)              
                cont += 1
                
        print(f'Puedes hacer {min(cantidades)} pasteles')
        result = min(cantidades) 
        
            
    else:
        print('Cuando tengas todos los ingredientes ¡nos ponemos a cocinar!')
        result = 0
    return result
````
