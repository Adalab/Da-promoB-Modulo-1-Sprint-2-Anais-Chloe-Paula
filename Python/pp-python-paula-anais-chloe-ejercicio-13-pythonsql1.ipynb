2. Empezemos a explorar la BBDD de Northwind. ¿Qué tablas componen la BBDD? 
Recuerda que primero nos tendremos que conectar con la BBDD sobre la que queremos trabajar. 
Guarda el resultado de los nombres de la tablas en una lista.

```python
# Volvemos a abrir conexión porque vamos a recoger el nombre de las tablas de otra BBDD, 'northwind'.

mydb = mysql.connector.connect(
              user='root', password='AlumnaAdalab',
              host='127.0.0.1',
              database='northwind'
)

mycursor = mydb.cursor()
mycursor.execute("SHOW TABLES")

lista_tablas = []
for i in mycursor:
    lista_tablas.extend(i)

print(lista_tablas)

mycursor.close()
mydb.close()

```

3. ¿Qué columnas tiene cada una de las tablas? Para esto tendréis que hacer un for loop para que nos saque el resultado de todas las columnas.

```python
# Volvemos a abrir conexión porque vamos a recoger el nombre de las columnas de cada tabla de northwind desde otra BBDD, 'information_schema'.

mydb = mysql.connector.connect(
              user='root', password='AlumnaAdalab',
              host='127.0.0.1',
              database='information_schema'
)

mycursor = mydb.cursor()

query_columnas = "SELECT TABLE_NAME, COLUMN_NAME FROM COLUMNS WHERE TABLE_SCHEMA = 'northwind'"

mycursor.execute(query_columnas)

for x in mycursor:
    print(x)

mycursor.close()
mydb.close()
```
